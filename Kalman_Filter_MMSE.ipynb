```
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```
measurements=[]
motions=[]
x=1
for i in range(1000):
    measurements.append(x/2+(25*x)/(1+x**2)+8*cos(1.2*i))
    motions.append((x**2)/20)
    x=x/2+(25*x)/(1+x**2)+8*cos(1.2*i)
```

```
def update(mean1, var1, mean2, var2):
    newMean = (var2*mean1 + var1*mean2)/(var2+var1)
    newVariance = 1/(1/var2 + 1/var1)
    return [newMean, newVariance] #returns updated gaussian parameters
```

```
def predict(mean1, var1, mean2, var2):
    new_mean = mean1 + mean2
    new_var = var1 + var2
    return [new_mean, new_var] #returns updated gaussian parameters after motion.

# Mean and covariances.
measurement_sig = 10.
motion_sig = 1.
mu = 0.
sig = 1000.

for k in range(len(measurements)):
    # measurement update with uncertainty
    mu, sig = update(mu, sig, measurements[k], measurement_sig)
    print('Update: [{}, {}]'.format(mu, sig))
    # motion update, with uncertainty
    mu, sig = predict(mu, sig, motions[k], motion_sig)
    print('Predict: [{}, {}]'.format(mu, sig))

print('\n')
print('Final result: [{}, {}]'.format(mu, sig))
```
